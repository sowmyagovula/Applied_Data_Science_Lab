## Clean `df2` and `df3`

```python
df2.head()
```

```python
df3.head()
```

```python
df2.drop(columns=["Unnamed: 0", "operation", "currency", "price", "surface_total_in_m2", "price_aprox_local_currency","price_usd_per_m2", "price_per_m2", "floor", "rooms", "expenses", "properati_url"], inplace= True)
df3.drop(columns=["Unnamed: 0", "operation", "currency", "price", "surface_total_in_m2", "price_aprox_local_currency","price_usd_per_m2", "price_per_m2", "floor", "rooms", "expenses", "properati_url"], inplace= True)
```

```python
df2.dropna(inplace= True)
df3.dropna(inplace= True)
```

```python
df2.isnull().sum()
```

```python
df3.isnull().sum()
```

```python
df2[['lat', 'lon']] = df2['lat-lon'].str.split(',', expand=True)
df3[['lat', 'lon']] = df3['lat-lon'].str.split(',', expand=True)
```

```python
df2['lat'] = df2['lat'].astype(float)
df2['lon'] = df2['lon'].astype(float)
```

```python
df3['lat'] = df3['lat'].astype(float)
df3['lon'] = df3['lon'].astype(float)
```

```python
df2.drop(columns=["lat-lon"], inplace = True)
df3.drop(columns=["lat-lon"], inplace = True)
```

```python
df2["state"] = df2['place_with_parent_names'].str.split('|').str[2]
df3["state"] = df3['place_with_parent_names'].str.split('|').str[2]
```

```python
df2.drop(columns=["place_with_parent_names"], inplace = True)
df3.drop(columns=["place_with_parent_names"], inplace = True)
```

```python
df1.head()
```

```python
df2.head()
```

```python
df3.head()
```

```python
# Concatenate df1, df2, and df3
df = pd.concat([df1,df2,df3])

# Print object type, shape, and head
print("df type:", type(df))
print("df shape:", df.shape)
df.head()
```

```python
df.info()
```

**Analysis of surface_covered_in_m2**

```python
df_trimmed = df[df["surface_covered_in_m2"] <= df["surface_covered_in_m2"].quantile(0.99)]

plt.figure(figsize=(8, 5))
plt.hist(df_trimmed["surface_covered_in_m2"], bins=50, color='lightgreen', edgecolor='black', label="surface_covered_in_m2")
plt.xlabel("Surface Covered (m²)")
plt.ylabel("Count")
plt.title("Distribution by sizes vs count (99th percentile trimmed)")
mn = np.mean(df["surface_covered_in_m2"])
plt.plot([mn for i in range(0,2000)], [i for i in range(0,2000)], "r--", label="mean of surface_covered_in_m2")
plt.legend()
plt.show()
```

By this we can say, the count of small houses are larger. Even though count of large houses is less but size is greater the mean is pulled towards the large surface covered(m2). This leads to error due to outliers.

```python
plt.boxplot(df_trimmed["surface_covered_in_m2"], vert=False)
plt.show()
```

# Research Question 1

**Which state has the most expensive real estate market?**

```python
mean_price_by_state = df.groupby("state")["price_aprox_usd"].mean().sort_values(ascending= False)
mean_price_by_state
```

```python
mean_price_by_state.plot(
    kind="bar",
    xlabel="state",
    ylabel="price",
    title="Mean House Price by State"
);
```

```python
df["price_per_m2"] = df['price_aprox_usd']/df['surface_covered_in_m2']
```

```python
df["price_per_m2"].unique()
```

```python
df[df["surface_covered_in_m2"]==0]
```

```python
df.shape
```

```python
df = df[df["surface_covered_in_m2"] != 0]
```

```python
df.shape
```

```python
df.head()
```

```python
df.groupby("state")["price_per_m2"].mean().sort_values(ascending= False)
```

```python
df.groupby("state")["price_per_m2"].mean().sort_values(ascending= False).plot(kind="bar", ylabel="mean price", title="Mean House Price per m2 by State");
```

By this we get to see Solidaridad and Quintana Roo are Top than Federal Distrito:

**Tourism and Luxury Real Estate in Solidaridad and Quintana Roo**

➊ Solidaridad includes Playa del Carmen, and Benito Juárez includes Cancún — two of the most visited international beach destinations in Mexico.

➋ These areas have a strong demand for luxury apartments, resorts, and vacation homes, often targeted at foreign buyers and wealthy investors.

**Distrito Federal (CDMX) has a more diverse housing market:**

➊ CDMX has a wide range of housing options — from upscale Polanco condos to lower-income areas — which brings the average price per m² down.

➋ While luxury areas in CDMX (like Roma, Condesa, Santa Fe) are expensive, many other neighborhoods have more affordable housing, creating a lower average.
